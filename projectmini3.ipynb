```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\YESHWANTH\titanic dataset.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Sex', y='Survived', hue='Pclass', data=df, ci=None)
plt.title('Survival Rate by Sex and Class')
plt.show()
```

```python
# KDE plot: Age distribution by survival
plt.figure(figsize=(8,6))
sns.kdeplot(df[df['Survived']==1]['Age'], label='Survived', shade=True)
sns.kdeplot(df[df['Survived']==0]['Age'], label='Not Survived', shade=True)
plt.title('Age Distribution by Survival')
plt.xlabel('Age')
plt.show()
```

```python
# Box plot: Age by class
plt.figure(figsize=(8,6))
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age Distribution by Passenger Class')
```

```python
# Violin plot: Fare by class
plt.figure(figsize=(8,6))
sns.violinplot(x='Pclass', y='Fare', data=df)
plt.title('Fare Distribution by Class')
```

```python
# Swarm plot (if small enough): Age vs Fare with survival hue
plt.figure(figsize=(8,6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, alpha=0.6)
plt.title('Age vs Fare (Survival Highlighted)')
```

```python
# Heatmap of correlations
plt.figure(figsize=(6,5))
corr = df[['Survived','Age','Fare','Pclass']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
```
